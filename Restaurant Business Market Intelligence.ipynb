```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Professional styling for all business reports
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Environment Setup: Success. Analytical libraries loaded.")
```

```python
# Simulating a high-quality restaurant dataset for market analysis
data = {
    'Restaurant_Name': ['Pizza Hut', 'Burger King', 'Subway', 'Local Dhaba', 'KFC', 'Dominoes', 'Starbucks', 'The Social', 'Chai Point', 'Haldirams'],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Delhi', 'Mumbai'],
    'Average_Cost_for_Two': [850, 500, 450, 250, 650, 750, 1100, 1600, 350, 600],
    'Total_Votes': [1800, 2400, 1200, 700, 2100, 2900, 3500, 4800, 950, 5500],
    'Aggregate_Rating': [4.2, 4.1, 3.8, 4.4, 3.9, 4.2, 4.5, 4.7, 3.9, 4.6]
}

df = pd.DataFrame(data)
print("Data Acquisition Complete: Market data loaded into DataFrame.")
display(df.head())
```

```python
# Logic: Success = High Rating + High Customer Engagement (Votes)
df['Market_Popularity'] = (df['Aggregate_Rating'] * df['Total_Votes']) / 1000

# Logic: Segmenting restaurants based on investment (Price Category)
def business_segment(cost):
    if cost <= 400: return 'Budget'
    elif cost <= 800: return 'Mid-Range'
    else: return 'Premium'

df['Segment'] = df['Average_Cost_for_Two'].apply(business_segment)

print("Feature Engineering: New business metrics defined.")
display(df[['Restaurant_Name', 'Segment', 'Market_Popularity']].sort_values(by='Market_Popularity', ascending=False))
```

```python
plt.figure(figsize=(9, 5))
# Using a bar chart to show distribution
sns.countplot(x='Segment', data=df, palette='coolwarm', order=['Budget', 'Mid-Range', 'Premium'])

plt.title('Market Strategy: Distribution of Restaurants by Segment', fontsize=15, fontweight='bold')
plt.xlabel('Market Segment Type', fontsize=12)
plt.ylabel('Count of Outlets', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# Bubble Chart: Bubble size indicates customer footfall (Total Votes)
sns.scatterplot(x='Average_Cost_for_Two', y='Market_Popularity', 
                size='Total_Votes', hue='Segment', 
                data=df, sizes=(200, 2000), alpha=0.6)

plt.title('Business Matrix: Investment (Cost) vs. Market Reach (Popularity)', fontsize=15, fontweight='bold')
plt.xlabel('Average Cost for Two (Investment)', fontsize=12)
plt.ylabel('Popularity Index (Business Performance)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, title="Business Segment")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
